```python
%matplotlib widget
from utils.algorithms.distributions import clustering_benchmark_suite
from utils.algorithms.clustering import k_means, expect_max
from utils.clustering_viz import visualize_clustering_suite


def expect_max_wrapper(X, K=2, steps=20):
    for X_c, theta in expect_max(X, K=K, steps=steps):
        means, _, _ = theta
        yield X_c, means


datasets = clustering_benchmark_suite(n_samples=400, random_state=None)

solvers = {
    "K-Means": k_means,
    "Expectation-Maximization": expect_max_wrapper,
}

visualize_clustering_suite(
    datasets=datasets,
    solvers=solvers,
    figsize=(12, 10),
)
```
