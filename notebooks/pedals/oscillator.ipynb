# op-amp schmitt trigger relaxation oscillator

Wir verwenden für den Feedback-Widerstand einen Widerstand $R_c$ in serie mit einem potentiometer $R_f$ und einen On/Off/On switch der $C_2$ oder $C_3$ zusätzlich parallel zu $C_1$ schaltet. So kann man zwischen drei verschiedenen groben ranges (sub audio, low audio, high audio) und mit dem poti zusätzlich fein einstellen.

Für einen opamp schmitt trigger gelten die treshold spannungen

$$\begin{aligned}U_{th+} &= \frac{R_1}{R_1+R_2} U_\text{high}\\ 
U_{th-} &= \frac{R_1}{R_1+R_2} U_\text{low}\end{aligned}$$ 

Da wir nicht mit symmetrischen Versorgungsspannungen, sondern mit $V_{EE} = 0V$ und $V_{CC} = +9V$ arbeiten, ergibt sich

$$\begin{aligned}U_{th+} &= \frac{R_1}{R_1+R_2} U_\text{high} + \frac{R_2}{R_1+R_2} 4.5V\\
U_{th-} &= \frac{R_1}{R_1+R_2} U_\text{low} + \frac{R_2}{R_1+R_2} 4.5V \end{aligned} $$ 

Entsprechend verwenden wir auch ein virtual ground bei $4.5V$, das wir bei den Kondensatoren sowie dem $R_1$/$R_2$ spannungsteiler statt dem normalen ground verwenden. Die opamp rails verwenden weiter das normale ground bei $0V$

---

Bei idealen opamps kann man zur Berechnung der Frequenz die Formel

$$
f=\frac{1}{2RC\ln\frac{1+\beta}{1-\beta}}, \beta = \frac{R_1}{R_1+R_2}
$$
verwenden. Diese Formel nimmt an, dass der opamp rail-to-rail swingt. Für reale opamps berechnet man erst die zeitintervalle, in denen der output LOW bzw. HIGH ist und berechnet dann den Kehrwert der Summe dieser.

Unter der Annahme dass der eingangswiderstand des Op-amps $\rightarrow\infty$ kann man den lade- bzw. entladestrom als den Strom durch den Kondensator berechnen, also $I_\text{charge} = \frac{U_\text{high}-U_C}{R}$ und $I_\text{discharge} = \frac{U_\text{low}-U_C}{R}$. In realen opamps ist neben der Eingangsimpedanz auch die Slew Rate relevant, die ignoriere ich hier, weil sie für die hier relevanten Frequenzen vernachlässigbar ist.

Weiter ist bekannt, dass $U_C(t) = \frac{1}{C} \int_0^t i(t) dt$. Demnach lässt sich eine entsprechende Differentialgleichung aufstellen
$$
\begin{aligned}
i(t) &= \frac{U_\text{out}-U_C(t)}{R} \\
\frac{\partial U_C(t)}{\partial t} &= \frac{U_\text{out}-U_C(t)}{CR} \\
\Rightarrow U_C(t) &= U_\text{out} - (U_\text{out} - U_{C,0}) e^{-t/RC} 
\end{aligned}
$$
$U_{out}$ entspricht hier jeweils $U_\text{high}$ bzw $U_\text{low}$ und $U_{C,0}$ entsprechend $U_{th-}$ und $U_{th+}$, je nach dem ob die Formel für HIGH oder LOW gesucht ist. Die Zeit, die der opamp jeweils auf LOW bzw HIGH liegt, entspricht der Zeit, bei der $U_C(t) = U_{th+}$ (bzw $-$) ist. Man kann dies also einsetzen und entsprechend umformen.

$$
\begin{aligned}
U_{th+} &= U_\text{high} - (U_\text{high}-U_{th-}) e^{t_\text{high}/RC} \\
t_\text{high} &= RC \ln\frac{U_\text{high}-U_{th-}}{U_\text{high} -V_{th+}} \\ \\
U_{th-} &= U_\text{low} - (U_\text{low} -U_{th+}) e^{t_\text{low}/RC} \\
t_\text{low} &= RC \ln\frac{U_\text{low} -U_{th+}}{U_\text{low} -U_{th-}} 
\end{aligned}
$$

Und für die Frequenz gilt dann eben $f = (t_\text{high}+t_\text{low})^{-1}$, oder

$$\begin{aligned}f &= \frac{1}{t_\text{high}+t_\text{low}} \\
&=\frac{1}{RC \ln\frac{U_\text{high}-U_{th-}}{U_\text{high} -V_{th+}}+RC \ln\frac{U_\text{low} -U_{th+}}{U_\text{low} -V_{th-}}} \\
&=\frac{1}{RC\left[\ln\frac{U_\text{high}-U_{th-}}{U_\text{high} -V_{th+}}+\ln\frac{U_\text{low} -U_{th+}}{U_\text{low} -V_{th-}}\right]} \\
&=\frac{1}{RC\ln\left(\frac{U_\text{high}-U_{th-}}{U_\text{high} -V_{th+}}\cdot\frac{U_\text{low} -U_{th+}}{U_\text{low} -V_{th-}}\right)} 
\end{aligned}$$

mit $U_{th+} = \beta U_\text{high}+(1-\beta) 4.5V$, $U_{th-} = \beta U_\text{low} + (1-\beta) 4.5V$, $\beta = \frac{R_1}{R_1+R_2}$
$$
\begin{aligned}&=\frac{1}{RC\ln\left(\frac{U_\text{high}-(\beta U_\text{low} + (1-\beta) 4.5V)}{U_\text{high} -(\beta U_\text{high}+(1-beta) 4.5V)}\cdot\frac{U_\text{low} -(\beta U_\text{high}+(1-beta) 4.5V)}{U_\text{low} -(\beta U_\text{low} + (1-\beta) 4.5V)}\right)} \\
&=\frac{1}{RC\ln\left(\frac{U_\text{high}-\beta U_\text{low} - (1-\beta) 4.5V}{U_\text{high} -\beta U_\text{high}-(1-\beta) 4.5V}\cdot\frac{U_\text{low} -\beta U_\text{high}-(1-\beta) 4.5V}{U_\text{low} -\beta U_\text{low} - (1-\beta) 4.5V}\right)} 
\end{aligned}
$$
$R$ entspricht $R_f + R_c$ und $C$ entspricht je nach Konfiguration $C_1, C_1+C_2$ oder $C_1+C_3$

---

Die tatsächlichen Werte für $U_\text{high}$ und $U_\text{low}$ sind nicht nur nicht einfach die Versorgungsspannungen, sondern hängen auch von dem Lastwiderstand ab. Eine genaue Formel ist unmöglich zu finden und eine Annährerung ist auch nicht genau genug. Die Lösung hier ist zwischen ausgang des opamps und den Feedbackwiderständen einen Impedanzwandler einzufügen.

Jetzt ist es nur noch relevant, einmalig durch messung statige werte für $U_\text{high}$ und $U_\text{low}$ bei festen $R_1, R_2$ zu finden.

Für das optimisierungsproblem stellen wir erstmal fest:
- Werte für $R_c$ sollten zwischen $100\Omega$ und $1M\Omega$ liegen.
- Werte für die Kondensatoren sollten zwischen $100pF$ und $100\mu F$ liegen.
- Werte für den Poti sollten nur welche sein, für die es auch potis gibt
- Die Frequenzen werden jeweils für $R_f = 0$ und $R_f = R_{f,\text{max}}$ getestet werden
- Die Frequenzen sollten den selben geometrischen abstand haben, um gute ergebnisse zu erhalten. Im Skript wird dazu `np.geomspace(0.2, 6800, 4)` verwendet um 4 Punkte zwischen $0.2Hz$ und $6.8kHz$ zu erzeugen. 

Wir erhalten die Gleichungen
$$\begin{aligned}
0.2Hz &= \left((C_1+C_3)(R_f+R_c)\ln(\dots)\right)^{-1} \\
6.4Hz &= \left((C_1+C_3)R_c\ln(\dots)\right)^{-1} \\
6.4Hz &= \left((C_1+C_2)(R_f+R_c)\ln(\dots)\right)^{-1} \\
210Hz &= \left((C_1+C_2)R_c\ln(\dots)\right)^{-1} \\
210Hz &= \left(C_1(R_f+R_c)\ln(\dots)\right)^{-1} \\
6.8kHz &= \left(C_1R_c\ln(\dots)\right)^{-1} \\
\end{aligned}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

base = np.geomspace(0.2, 6800, 4)
targets = np.array([base[0], base[1], base[1], base[2], base[2], base[3]])
log_targets = np.log(targets)

U_HIGH = 8.06
U_LOW = 0.932
V_REF = 4.5
BETA = 0.5

RC_MIN, RC_MAX = 100.0, 1_000_000.0
C_MIN, C_MAX = 100e-12, 100e-6

# Potentiometer end values: either bypassed (0) or one of these maxima
RF_OPTIONS = np.array([0.0, 50e3, 100e3, 250e3, 470e3, 1e6])

# Deterministic 1D search over Rc for each discrete Rf
COARSE_POINTS = 400
REFINE_POINTS = 280

def oscillator_k(u_high=U_HIGH, u_low=U_LOW, beta=BETA, v_ref=V_REF):
    uth_plus = beta * u_high + (1.0 - beta) * v_ref
    uth_minus = beta * u_low + (1.0 - beta) * v_ref

    term_high = (u_high - uth_minus) / (u_high - uth_plus)
    term_low = (u_low - uth_plus) / (u_low - uth_minus)

    if term_high <= 0 or term_low <= 0:
        raise ValueError("Invalid constants for logarithm in oscillator formula.")

    return np.log(term_high) + np.log(term_low)

def solve_cap_term_pair(r_min, r_max, f_min_target, f_max_target, k_const):
    x1 = -np.log(k_const * r_min) - np.log(f_min_target)
    x2 = -np.log(k_const * r_max) - np.log(f_max_target)
    return np.exp(0.5 * (x1 + x2))

def evaluate_candidate(rc, rf, k_const):
    # Each range uses R_total=Rc+Rf for low endpoint and R_total=Rc for high endpoint.
    r_low = rc + rf
    r_high = rc

    s1 = solve_cap_term_pair(r_low, r_high, targets[0], targets[1], k_const)
    s2 = solve_cap_term_pair(r_low, r_high, targets[2], targets[3], k_const)
    s3 = solve_cap_term_pair(r_low, r_high, targets[4], targets[5], k_const)

    c1 = s3
    c2 = s2 - s3
    c3 = s1 - s3

    if not (C_MIN <= c1 <= C_MAX and C_MIN <= c2 <= C_MAX and C_MIN <= c3 <= C_MAX):
        return None

    cap_terms = np.array([c1 + c3, c1 + c3, c1 + c2, c1 + c2, c1, c1])
    r_terms = np.array([r_low, r_high, r_low, r_high, r_low, r_high])
    vals = 1.0 / (k_const * cap_terms * r_terms)

    err = np.mean((np.log(vals) - log_targets) ** 2)
    return c1, c2, c3, vals, err

def refine_grid(center, lo, hi, points, span_decades=0.55):
    left = max(lo, center / (10.0 ** span_decades))
    right = min(hi, center * (10.0 ** span_decades))
    if left >= right:
        return np.array([center])
    return np.geomspace(left, right, points)

def is_better(err, rc, rf, best_err, best_rc, best_rf):
    if err < best_err:
        return True
    if not np.isclose(err, best_err):
        return False
    if rf < best_rf:
        return True
    if np.isclose(rf, best_rf) and rc < best_rc:
        return True
    return False

def run():
    k_const = oscillator_k()

    rc_coarse = np.geomspace(RC_MIN, RC_MAX, COARSE_POINTS)
    total = len(RF_OPTIONS) * (len(rc_coarse) + REFINE_POINTS)
    progress = widgets.IntProgress(value=0, min=0, max=total, description="Search")
    status = widgets.HTML(value="")
    chart_out = widgets.Output()
    display(widgets.VBox([progress, status, chart_out]))

    history = []
    best_err = np.inf
    best_solution = None
    step = 0

    best_rc_by_rf = {}

    # Stage 1: global deterministic scan over discrete Rf options and Rc grid
    for rf in RF_OPTIONS:
        best_err_rf = np.inf
        best_rc_rf = None

        for rc in rc_coarse:
            res = evaluate_candidate(rc, rf, k_const)
            if res is not None:
                c1, c2, c3, vals, err = res

                if err < best_err_rf or (np.isclose(err, best_err_rf) and (best_rc_rf is None or rc < best_rc_rf)):
                    best_err_rf = err
                    best_rc_rf = rc

                if best_solution is None or is_better(err, rc, rf, best_err, best_solution[3], best_solution[4]):
                    best_err = err
                    best_solution = (c1, c2, c3, rc, rf, vals)

            step += 1
            progress.value = step
            history.append(best_err if np.isfinite(best_err) else np.nan)

            if step % 1200 == 0:
                status.value = f"<b>Coarse scan</b> | <b>Step:</b> {step} / {total} | <b>Best log-MSE:</b> {best_err:.3e}"
                with chart_out:
                    clear_output(wait=True)
                    plt.figure(figsize=(7, 3))
                    plt.semilogy(history)
                    plt.xlabel("Step")
                    plt.ylabel("Best log-MSE")
                    plt.title("Convergence")
                    plt.grid(True, alpha=0.3)
                    plt.show()

        best_rc_by_rf[rf] = best_rc_rf

    if best_solution is None:
        raise RuntimeError("No valid solution found within the requested Rc/C ranges and Rf options.")

    # Stage 2: local deterministic refinement in Rc for each discrete Rf
    for rf in RF_OPTIONS:
        center_rc = best_rc_by_rf[rf]
        if center_rc is None:
            # Keep progress deterministic even if this branch has no feasible coarse point.
            step += REFINE_POINTS
            progress.value = step
            continue

        rc_refine = refine_grid(center_rc, RC_MIN, RC_MAX, REFINE_POINTS, span_decades=0.5)

        for rc in rc_refine:
            res = evaluate_candidate(rc, rf, k_const)
            if res is not None:
                c1, c2, c3, vals, err = res
                if is_better(err, rc, rf, best_err, best_solution[3], best_solution[4]):
                    best_err = err
                    best_solution = (c1, c2, c3, rc, rf, vals)

            step += 1
            progress.value = step
            history.append(best_err)

            if step % 1200 == 0 or step == total:
                status.value = f"<b>Refine scan</b> | <b>Step:</b> {step} / {total} | <b>Best log-MSE:</b> {best_err:.3e}"
                with chart_out:
                    clear_output(wait=True)
                    plt.figure(figsize=(7, 3))
                    plt.semilogy(history)
                    plt.xlabel("Step")
                    plt.ylabel("Best log-MSE")
                    plt.title("Convergence")
                    plt.grid(True, alpha=0.3)
                    plt.show()

    c1b, c2b, c3b, rcb, rfb, valsb = best_solution

    print("Best solution (new formula, discrete potentiometer options)")
    print(f"U_high={U_HIGH:.3f} V, U_low={U_LOW:.3f} V, beta={BETA:.3f}, Vref={V_REF:.3f} V")
    print(f"K={k_const:.6f}")
    print(f"Selected Rf_max={rfb:.1f} ohm (frequencies evaluated at R_total=Rc and Rc+Rf_max)")
    print(f"C1={c1b:.3e} F ({c1b * 1e12:.1f} pF)")
    print(f"C2={c2b:.3e} F ({c2b * 1e9:.3f} nF)")
    print(f"C3={c3b:.3e} F ({c3b * 1e6:.3f} uF)")
    print(f"Rc={rcb:.1f} ohm")
    print(f"log-MSE={best_err:.6e}\n")

    print("eq   target      achieved    error_%")
    for i, (t, v) in enumerate(zip(targets, valsb), 1):
        e = 100.0 * (v - t) / t
        print(f"{i:<4}{t:<12.6g}{v:<12.6g}{e:+.2f}")

    plt.figure(figsize=(7, 3.2))
    x = np.arange(len(targets))
    w = 0.38
    plt.bar(x - w / 2, targets, width=w, label="Target")
    plt.bar(x + w / 2, valsb, width=w, label="Achieved")
    plt.yscale("log")
    plt.xticks(x, ["fmin1", "fmax1", "fmin2", "fmax2", "fmin3", "fmax3"])
    plt.ylabel("Frequency (Hz)")
    plt.title("Target vs Achieved")
    plt.legend()
    plt.grid(axis="y", alpha=0.3)
    plt.show()

run()
```
