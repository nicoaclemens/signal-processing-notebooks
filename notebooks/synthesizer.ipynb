## Controls

### Controllers

- KNOB: Tune (-2, 2) (semitones)
- KNOB: Glide (0, 10) (portamento time)
- KNOB: Modulation Mix (0, 10); Label at 0: "OSC. 3/\nFilter EG"; Label at 10: "Noise/LFO"
- GRAY BUTTON, Horizontal: Left: Osc. 3, Right: Filter EG (Use Osc. 3 or Filter Envelope Generator as modulation source)
- GRAY BUTTON, Horizontal: Left: NOISE, Right: LFO (Use noise generator or lfo; feeds into modulation mix knob)

### Oscillator Bank

- ORANGE BUTTON, Horizontal Across the border to Controllers: Oscillator Modulation On/Off (global mod enable switch)
- Per Oscillator:
  - KNOB (DISCRETE): Range [LO, 32', 16', 8', 4', 2']
  - KNOB (DISCRETE): Waveform [TRIANGLE, Triangle + Saw (sharktooth), RAMP, PWM1, PWM2, PWM3]
- Oscillator 2, 3:
  - KNOB: Tune (-7, 7)
- ORANGE BUTTON, Vertical Osc. 3 Controll (turns keyboard control for osc. 3 on/off; if off, osc. 3 is a fixed frequency)

unsure what duty cylcles PWM1, PWM2, PWM3 have. Assuming 50%, 25%, 12%; They are labeled "Square, Wide, Narrow" in the user manual. 

### Mixer

- Per Oscillator, aligned with the ones from Oscillator Bank:
  - KNOB: Volume (0, 10)
  - CYAN SWITCH: On/Off
- External Input section (ignored):
  - Aux In
  - KNOB: Volume (0, 10)
  - CYAN SWITCH: On/Off
- Noise Section:
  - CYAN SWITCH: On/Off
  - KNOB: Volume (0, 1)
  - CYAN SWITCH: White/Pink

All Switches except White/Pink horizontal

### Modifiers 

all Attack & Decay time Knobs in the format:
min label: "m-sec."
max label: "sec."
numbered labels (in that order from min to max): 10, (empty), 200, (empty), 600, (empty), 1, (empty), 5, (empty), 10
#### Filter

- ORANGE VERTICAL BUTTON, on border to Mixer:
  - Filter Modulation
  - Keyboard Control 1
  - Keyboard Control 2

- KNOB: Cutoff Frequency (-4, 4)
- KNOB: Attack Time 
- KNOB: Emphasis (0, 10)
- KNOB: Decay Time
- KNOB: Amount of Contour (0, 10)
- KNOB: Sustain Level (0, 10)

#### Loudness Contour
- KNOB: Attack Time
- KNOB: Decay Time
- KNOB: Sustain Level (0. 10)

### Output

- Knob: Volume
- CYAN BUTTON: Horizontal, "Main Output", On/Off
- CYAN BUTTON: Horitontal, "A=440", On/Off
- KNOB: Volume (ignored)
- AUX JACK (PHONES) (ignored)

### LFO Controller
- KNOB: LFO Rate
- WHITE BUTTON: Horizontal, "Glide", On/Off
- WHITE BUTTON: Horizontal, "Decay", On/Off
- Wheel (Slider): Vertical; "Pitch"
- Wheel (Slider): Vertical; "Mod."
