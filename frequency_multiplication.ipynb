# Frequency Multiplcation Visualiser



Die Multiplikation zweier Sinussignale erzeugt ein Signal, welches die Summe und Differenz der beiden Frequenzen als Frequenzanteile hat. Es gilt die allgemeine Formel

$$\cos\alpha \cos \beta = \frac{1}{2}\left[\cos(\alpha+\beta)+\cos(\alpha-\beta)\right]$$

Diese Identität lässt sich leicht mit Euler's Formel $\cos(x) = \frac{1}{2}\left(e^{ix}+e^{-ix}\right)$ beweisen

$$\begin{aligned}\cos\alpha\cos\beta &= \frac{e^{i\alpha}+e^{-i\alpha}}{2} \frac{e^{i\beta}+e^{-i\beta}}{2} \\
&=\frac{\left(e^{i\alpha}+e^{-i\alpha}\right)\left(e^{i\beta}+e^{-i\beta}\right)}{4} \\
&=\frac{e^{i\alpha+i\beta}+e^{i\alpha-i\beta}+e^{-i\alpha+i\beta}+e^{-i\alpha-i\beta}}{4} \\
&=\frac{e^{i(\alpha+\beta)}+e^{i(\alpha-\beta)}+e^{-i(\alpha+\beta)}+e^{-i(\alpha-\beta)}}{4} \\
&=\frac{1}{2}\left[\frac{e^{i(\alpha+\beta)}+e^{-i(\alpha+\beta)}}{2}+\frac{e^{i(\alpha-\beta)}+e^{-i(\alpha-\beta)}}{2}\right] \\
&=\frac{1}{2}\left[\cos(\alpha+\beta)+\cos(\alpha-\beta)\right]
\end{aligned}
$$

Wenn $f_1 \gg f_2$, dann ist das Ausgangssignal eine Amplitudenmodulierte Version von $f_1$. Das bedeutet bspw in Musik ($f_1 \in [20, 8k], f_2 \in[0.1, 10]$) ein Tremolo-Effekt (Lautstärke wird langsam leiser und lauter) und in der Radiotechnik Amplitudenmodulation (AM); Dabei ist $f_1$ die Carrierfrequenz (typischerweise im Megahertz bereich) und $f_2$ die Modulierfrequenz (für Radio im Audiobereich). Im Frequenzspektrum hat man hier zwei Frequenzen sehr nah am Ausgangssignal

Wenn $f_1 \approx f_2$ hat man eine art "Ring-Modulation". 

Bei $f_1 = f_2$ ergeben sich für die Frequenzanteile $0$ und $2f_1$. Das Ergebis ist also eine frequenzdopplung mit einer Offset-spannung. In den meisten Systemen wird der Gleichanteil rausgefilert, das Ergebnis der Quadrierung einer Sinusschwingung ist also eine frequenzdopplung
